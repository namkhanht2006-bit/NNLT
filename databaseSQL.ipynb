# [dbo].[sp_KhoaBangDiem]

```sql
/****** Object:  StoredProcedure [dbo].[sp_KhoaBangDiem]    Script Date: 4/28/2026 8:13:40 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- Thủ tục nội tại
--1.GV khóa điểm sau khi nhập
CREATE PROCEDURE [dbo].[sp_KhoaBangDiem]
    @MaPC VARCHAR(20)
AS
BEGIN
    -- Kiểm tra xem lớp phân công này có tồn tại không
    IF NOT EXISTS (SELECT 1 FROM PhanCong WHERE MaPC = @MaPC)
    BEGIN
        RAISERROR(N'Mã phân công không tồn tại!', 16, 1);
        RETURN;
    END

    -- Thực hiện khóa điểm (chuyển TrangThaiKhoa từ 0 -> 1)
    UPDATE Diem
    SET TrangThaiKhoa = 1
    WHERE MaPC = @MaPC AND TrangThaiKhoa = 0;

    PRINT N'Đã khóa bảng điểm thành công cho mã phân công: ' + @MaPC;
END;
GO
```

# [dbo].[sp_NopDonPhucKhao]

```sql
/****** Object:  StoredProcedure [dbo].[sp_NopDonPhucKhao]    Script Date: 4/28/2026 8:13:40 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
--2.HV nộp đơn phúc khảo
CREATE PROCEDURE [dbo].[sp_NopDonPhucKhao]
    @MaPK VARCHAR(20),
    @MaHV VARCHAR(20),
    @MaHP VARCHAR(20),
    @LyDo NVARCHAR(MAX)
AS
BEGIN
    DECLARE @DiemCu FLOAT;

    -- Tìm điểm thi cũ của học viên đối với học phần này
    SELECT TOP 1 @DiemCu = d.DiemThi
    FROM Diem d
    INNER JOIN PhanCong pc ON d.MaPC = pc.MaPC
    WHERE d.MaHV = @MaHV AND pc.MaHP = @MaHP;

    -- Nếu không tìm thấy điểm (chưa có điểm)
    IF @DiemCu IS NULL
    BEGIN
        RAISERROR(N'Học viên chưa có điểm môn này, không thể phúc khảo!', 16, 1);
        RETURN;
    END

    -- Thêm đơn phúc khảo với điểm cũ vừa lấy được
    INSERT INTO PhucKhao (MaPK, MaHV, MaHP, LyDo, DiemCu, TrangThaiXuLy)
    VALUES (@MaPK, @MaHV, @MaHP, @LyDo, @DiemCu, N'Chờ xử lý');

    PRINT N'Đã ghi nhận đơn phúc khảo!';
END;
GO
```

# [dbo].[sp_ThemViPhamKyLuat]

```sql
/****** Object:  StoredProcedure [dbo].[sp_ThemViPhamKyLuat]    Script Date: 4/28/2026 8:13:40 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
--4.Bổ sung vi phạm kỷ luật vào Hồ sơ rèn luyện
CREATE PROCEDURE [dbo].[sp_ThemViPhamKyLuat]
    @MaHV VARCHAR(20),
    @LoiViPham NVARCHAR(MAX),
    @MaCB_ChamDiem VARCHAR(20)
AS
BEGIN
    -- Kiểm tra học viên đã có hồ sơ chưa, nếu chưa có thì báo lỗi
    IF NOT EXISTS (SELECT 1 FROM HoSoRenLuyen WHERE MaHV = @MaHV)
    BEGIN
        RAISERROR(N'Học viên này chưa được khởi tạo Hồ sơ rèn luyện!', 16, 1);
        RETURN;
    END

    -- Cập nhật nối thêm chuỗi lỗi vi phạm vào trường KyLuat (xuống dòng bằng CHAR(13))
    UPDATE HoSoRenLuyen
    SET KyLuat = ISNULL(KyLuat, '') + CHAR(13) + N'- ' + CAST(GETDATE() AS VARCHAR) + N': ' + @LoiViPham,
        MaCB_ChamDiem = @MaCB_ChamDiem -- Ghi nhận người xử lý
    WHERE MaHV = @MaHV;

    PRINT N'Đã cập nhật lỗi vi phạm vào hồ sơ của học viên ' + @MaHV;
END;
GO
```

# [dbo].[sp_ThongKeNoTinChi]

```sql
/****** Object:  StoredProcedure [dbo].[sp_ThongKeNoTinChi]    Script Date: 4/28/2026 8:13:40 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
--3.Thống kê số tín chỉ nợ và Cảnh báo học vụ
CREATE PROCEDURE [dbo].[sp_ThongKeNoTinChi]
    @MaHV VARCHAR(20)
AS
BEGIN
    DECLARE @TongTinChiNo INT;

    -- Tính tổng số tín chỉ nợ dựa trên trạng thái 'Trượt'
    SELECT @TongTinChiNo = ISNULL(SUM(hp.SoTinChi), 0)
    FROM Diem d
    INNER JOIN PhanCong pc ON d.MaPC = pc.MaPC
    INNER JOIN HocPhan hp ON pc.MaHP = hp.MaHP
    WHERE d.MaHV = @MaHV AND d.TrangThai = N'Trượt';

    -- Đưa ra kết quả và cảnh báo
    IF (@TongTinChiNo >= 15)
    BEGIN
        PRINT N'CẢNH BÁO: Học viên ' + @MaHV + N' đang nợ ' + CAST(@TongTinChiNo AS VARCHAR) + N' tín chỉ. Vượt quá mức cho phép!';
        -- Ở đây có thể viết thêm lệnh UPDATE trạng thái cảnh báo vào bảng HocVien nếu có cột tương ứng
    END
    ELSE
    BEGIN
        PRINT N'Học viên ' + @MaHV + N' đang nợ ' + CAST(@TongTinChiNo AS VARCHAR) + N' tín chỉ. (Trong ngưỡng an toàn)';
    END
END;
GO
```
